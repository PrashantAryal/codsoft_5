## Data Preprocessing

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df[['Amount']])
```

```python
df.head()
```

### Handling Class Imbalance using SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
x = df.drop(columns=['Class','Time'])
y = df['Class']

smote = SMOTE(random_state=42)
x_resampled,y_resampled = smote.fit_resample(x,y)
```

```python
sns.countplot(x=y_resampled, palette='coolwarm')
plt.title('Class Distribution After SMOTE')
plt.xlabel('Class (0 = Genuine, 1 = Fraudulent)')
plt.ylabel('Count')
plt.show()
```

## model building and training

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, precision_recall_curve, auc
#from sklearn.metrics import plot_confusion_matrix

x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)

y_pred_lr = lr_model.predict(x_test)
y_score_lr = lr_model.predict_proba(x_test)[:,-1]

print("Logistic Regression Classification Report:\n")
print(classification_report(y_test, y_pred_lr))

roc_auc_lr = roc_auc_score(y_test, y_score_lr)
print(f"Logistic Regression ROC-AUC: {roc_auc_lr:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators = 10,random_state=42)
rf_model.fit(x_train, y_train)

y_pred_rf = rf_model.predict(x_test)
y_score_rf = rf_model.predict_proba(x_test)[:, 1]

print("Random Forest Classification Report:\n")
print(classification_report(y_test, y_pred_rf))

roc_auc_rf = roc_auc_score(y_test, y_score_rf)
print(f"Random Forest ROC-AUC: {roc_auc_rf:.4f}")
```

```python
from xgboost import XGBClassifier
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(x_train, y_train)

y_pred_xgb = xgb_model.predict(x_test)
y_score_xgb = xgb_model.predict_proba(x_test)[:, 1]

print("XGBoost Classification Report:\n")
print(classification_report(y_test, y_pred_xgb))

roc_auc_xgb = roc_auc_score(y_test, y_score_xgb)
print(f"XGBoost ROC-AUC: {roc_auc_xgb:.4f}")
```

```python
from sklearn.metrics import confusion_matrix
y_pred = rf_model.predict(x_test)  
cm = confusion_matrix(y_test, y_pred) 

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Genuine', 'Fraud'], yticklabels=['Genuine', 'Fraud'], cbar=False)

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Random Forest')

# Show the plot
plt.show()
```

```python
precision, recall, _ = precision_recall_curve(y_test, y_score_rf)
pr_auc = auc(recall, precision)

plt.plot(recall, precision, color='blue', label=f'AUPRC = {pr_auc:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```

```python
fpr, tpr, _ = roc_curve(y_test, y_score_rf)
roc_auc = roc_auc_score(y_test, y_score_rf)

plt.plot(fpr, tpr, color='blue', label=f'ROC-AUC = {roc_auc:.4f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
model_results = {
    'Model': ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'ROC-AUC': [roc_auc_lr, roc_auc_rf, roc_auc_xgb]
}

results_df = pd.DataFrame(model_results)

sns.barplot(x='ROC-AUC', y='Model', data=results_df, palette='coolwarm')
plt.title('Model Comparison')
plt.xlabel('ROC-AUC Score')
plt.ylabel('Model')
plt.show()
```
